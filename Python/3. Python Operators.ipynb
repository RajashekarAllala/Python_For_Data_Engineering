**Python Operators :**

Operators are used to perform operations on variables and values. Python divides the operators in the following groups:

* Arithmetic operators
* Assignment operators
* Comparison operators
* Logical operators
* Identity operators
* Membership operators
* Bitwise operators

**Python Arithmetic Operators :**
Arithmetic operators are used with numeric values to perform common mathematical operations.

| Operator |	Name |	Example|
|----------|---------|---------|
| +	| Addition	| x + y	|
| -	| Subtraction |	x - y |	
| *	| Multiplication |	x * y |	
| /	| Division |	x / y	|
| %	| Modulus |	x % y	|
| ** |	Exponentiation |	x ** y	|
| // |	Floor division	| x // y |

**Operator Precedence :**
* Operator precedence describes the order in which operations are performed.
* Parentheses has the highest precedence, meaning that expressions inside parentheses must be evaluated first.
* Multiplication * has higher precedence than addition +, and therefore multiplications are evaluated before additions.
* If two operators have the same precedence, the expression is evaluated from left to right.

```python
# Parentheses has the highest precedence, meaning that expressions inside parentheses must be evaluated first.
print("Testing Operator Precedence. Value after evaluation:", (9 + 3) - (6 + 3))

# Multiplication * has higher precedence than addition +
# Therefore multiplications are evaluated before additions.
print("Testing Operator Precedence. Value after evaluation:", 100 + 5 * 3)

# Addition + and subtraction - has the same precedence.
# Therefore we evaluate the expression from left to right.
print("Testing Operator Precedence. Value after evaluation:", 5 + 4 - 7 + 3)
```
