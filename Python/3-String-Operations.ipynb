Different String Operations

Accessing string characters using an index value<br>
Explaination: In Python, strings are indexed starting from 0.<br>
So, the first character of the string is at index 0, the second character is at index 1, and so on.<br>
You can access individual characters of a string using square brackets [] and the index value.

```python
my_string = "Hello, World!"
first_char = my_string[0]  # 'H'
print(f"First character: {first_char}")
```

String length<br>
Explaination: The len() function in Python is used to get the length of a string.<br>
It returns the number of characters in the string, including spaces and special characters.

```python
string_length = len(my_string)  # 13
print(f"String length: {string_length}")
```

Looping Through a String<br>
Explaination: You can loop through each character in a string using a for loop.

```python
print("Characters in the string:")
for x in my_string:
    print(x, end=' ')
```

Check if a substring exists within a string<br>
Explaination: The in operator is used to check if a substring exists within a string.<br>
It returns True if the substring is found, and False otherwise.

```python
substring = "World"
if substring in my_string:
    print(f"\n'{substring}' found in '{my_string}'")
else:
    print(f"\n'{substring}' not found in '{my_string}'")
```

Alternatively, using the find() method<br>
Explaination: The find() method returns the lowest index of the substring if it is found in the string.<br>
If the substring is not found, it returns -1.<br>
This method is case-sensitive.

```python
position = my_string.find(substring)
if position != -1:
    print(f"'{substring}' found at position {position}")
else:
    print(f"'{substring}' not found in '{my_string}'")
```

Using in operator with if statement<br>
Explaination: The in operator can be used in conditional statements to execute code based on the presence of a substring.<br>
Here, we check if "Hello" is present in the string.<br>
If it is, we print a message, otherwise we print another message.<br>
This is useful for validating input or checking conditions in your code.

```python
if "Hello" in my_string:
    print("'Hello' is present in the string")
else:
    print("'Hello' is not present in the string")
```

using not in operator with if statement<br>
Explaination: The not in operator is used to check if a substring does not exist within a string.<br>
It returns True if the substring is not found, and False if it is found.

```python
if "Python" not in my_string:
    print("'Python' is not present in the string")
else:
    print("'Python' is present in the string")
```

String Slicing with start and end indexes<br>
Explaination: String slicing allows you to extract a portion of a string by specifying the start and end indexes.<br>
The syntax for slicing is string[start:end], where start is the index of the first character<br>
you want to include in the slice and end is the index of the first character you want to exclude from the slice.<br>
The slice will include characters from the start index up to, but not including, the end index.<br>
If you omit the start index, it defaults to 0 (the beginning of the string).<br>
If you omit the end index, it defaults to the length of the string (the end of the string).<br>
You can also use negative indexes to slice from the end of the string.<br>
Negative indexes count from the end of the string, with -1 being the last character, -2 being the second to last character, and so on.<br>
You can also specify a step value to skip characters in the slice.<br>
The syntax for slicing with a step value is string[start:end:step], where step is the number of characters to skip.<br>
The step value can be positive or negative. A positive step value slices the string from left to right,<br>
while a negative step value slices the string from right to left.

```python
sliced_string = my_string[0:5]  # 'Hello'
print(f"Sliced string: {sliced_string}")
```

Slicing from the beginning

```python
sliced_string = my_string[:5]  # 'Hello'
print(f"Sliced string: {sliced_string}")
```

Slicing to the end

```python
sliced_string = my_string[7:]  # 'World!'
print(f"Sliced string: {sliced_string}")
```

Slicing with step value

```python
sliced_string = my_string[::2]  # 'Hlo ol!'
print(f"Sliced string with step: {sliced_string}")
```

Slicing with step value with start and end indexes

```python
sliced_string = my_string[0:13:2]  # 'Hlo ol!
print(f"Sliced string with start, end and step: {sliced_string}")
```

Negative Indexing

```python
last_char = my_string[-6]  # 'W'
print(f"Character using negative indexing: {last_char}")
```

Negative Indexing with slicing

```python
sliced_string = my_string[-6:-1]  # 'World'
print(f"Sliced string using negative indexing: {sliced_string}")
```

Negative Indexing with slicing and step value

```python
sliced_string = my_string[-1:-14:-2]  # '!lo olH'
print(f"Sliced string using negative indexing and step: {sliced_string}")
```

Reversing a string using slicing

```python
reversed_string = my_string[::-1]  # '!dlroW ,olleH'
print(f"Reversed string: {reversed_string}")
```

Replace a substring with another substring<br>
Explaination: The replace() method is used to replace occurrences of a specified substring with another substring.<br>
It takes two arguments: the substring to be replaced and the substring to replace it with.<br>
The method returns a new string with the replacements made.<br>
If the substring to be replaced is not found in the original string, the method returns the original string unchanged.<br>
The replace() method is case-sensitive, meaning that it will only replace substrings that match the case of the specified substring.<br>
You can also specify the maximum number of replacements to be made by providing an optional third argument.<br>
If the third argument is not provided, all occurrences of the substring will be replaced.<br>
If the third argument is provided, only the first n occurrences of the substring will be replaced.<br>
Here is an example:<br>
original_string = "Hello, World! Welcome to the World!"<br>
replaced_string = original_string.replace("World", "Python", 1)  #<br>
print(replaced_string)  # 'Hello, Python! Welcome to the World!'<br>
In this example, only the first occurrence of "World" is replaced with "Python".<br>
The second occurrence remains unchanged.<br>
If you want to replace all occurrences, you can omit the third argument:<br>
original_string = "Hello, World! Welcome to the World!"<br>
replaced_string = original_string.replace("World", "Python")  #<br>
print(replaced_string)  # 'Hello, Python! Welcome to the Python!'<br>
In this example, all occurrences of "World" are replaced with "Python".

```python
replaced_string = my_string.replace("World", "Python")  # 'Hello, Python!'
print(f"Replaced string: {replaced_string}")
```

print repeated words in a string<br>
Explaination: To find and print repeated words in a string, you can split the string into individual words using the split() method.<br>
Then, you can use a loop to iterate through the list of words and count the occurrences of each word using the count() method.<br>
If a word appears more than once, you can add it to a set to avoid duplicates.<br>
Finally, you can print the set of repeated words.

```python
words = string1.split()
repeated_words = set()
for word in words:
    word_count = string1.count(word)
    if word_count > 1:
        repeated_words.add(word)
print(f"Repeated words: {repeated_words}")
```

encode() method to encode a string<br>
Explaination: The encode() method is used to encode a string into a byte object.<br>
It takes one argument: the encoding format (default is 'utf-8').<br>
The method returns a byte object representing the encoded string.<br>
The encode() method is useful when you need to convert a string to bytes for storage or transmission.<br>
The most common encoding format is 'utf-8', which can represent all characters in the Unicode standard.<br>
You can also use other encoding formats, such as 'ascii', 'latin-1', or 'utf-16', depending on your requirements.<br>
If the specified encoding format is not supported, the method raises a LookupError.<br>
If the string contains characters that cannot be encoded in the specified format, the method raises a UnicodeEncodeError.<br>
To handle encoding errors, you can specify an optional error handling scheme as a second argument to the encode() method.<br>
The most common error handling schemes are 'strict' (default), 'ignore', and 'replace'.<br>
'strict' raises a UnicodeEncodeError on encoding errors, 'ignore' ignores the characters that cannot be encoded,<br>
and 'replace' replaces the characters that cannot be encoded with a replacement character (usually '?').

```python
encoded_string = my_string.encode("utf-8")  # b'Hello, World!'
print(f"Encoded string: {encoded_string}")
```

f-string for string formatting (Python 3.6+)<br>

To create an f-string, prefix the string with the letter 'f' or 'F'.<br>

They are a popular choice for string formatting in modern Python code.<br>

If you are using an earlier version of Python, you will need to use the format() method or other string formatting techniques.

String Formatting Types<br>
Explaination: Python provides several ways to format strings, including the use of format specifiers.<br>
Format specifiers are used to control the appearance of the formatted values, such as alignment, padding, number formatting, and more.<br>
Here are some commonly used format specifiers:

Using :< to left align<br>
Explaination: The :< format specifier is used to left align a string within a specified width.<br>
It takes one argument: the width of the resulting string.<br>
The method returns a new string that is left aligned within the specified width, with spaces added to the right if necessary.<br>
If the specified width is less than or equal to the length of the original string, the original string is returned unchanged.<br>
This format specifier is useful for creating tables or aligning text in a consistent way.

```python
left_aligned = "{:<10}".format("Hello")  # 'Hello     '
print(f"Left aligned: '{left_aligned}'")
```

Using :> to right align<br>
Explaination: The :> format specifier is used to right align a string within a specified width.<br>
It takes one argument: the width of the resulting string.<br>
The method returns a new string that is right aligned within the specified width, with spaces added to the left if necessary.<br>
If the specified width is less than or equal to the length of the original string, the original string is returned unchanged.<br>
This format specifier is useful for creating tables or aligning text in a consistent way.

```python
right_aligned = "{:>10}".format("Hello")  # '     Hello'
print(f"Right aligned: '{right_aligned}'")
```

Using :^ to center align<br>
Explaination: The :^ format specifier is used to center align a string within a specified width.<br>
It takes one argument: the width of the resulting string.<br>
The method returns a new string that is centered within the specified width, with spaces added to both sides if necessary.<br>
If the specified width is less than or equal to the length of the original string, the original string is returned unchanged.<br>
This format specifier is useful for creating tables or aligning text in a consistent way.

```python
center_aligned = "{:^10}".format("Hello")  # '  Hello   '
print(f"Centered aligned: '{center_aligned}'")
```

Using := to pad with zeros<br>
Explaination: The :0> format specifier is used to pad a string with leading zeros to a specified width.<br>
It takes one argument: the width of the resulting string.<br>
The method returns a new string that is padded with leading zeros to the specified width.<br>
If the specified width is less than or equal to the length of the original string, the original string is returned unchanged.<br>

Using := to keep the string to left most position<br>
Explaination: The := format specifier is used to pad a string with a specified character to a specified width.<br>
It takes two arguments: the character to pad with and the width of the resulting string.<br>
The method returns a new string that is padded with the specified character to the specified width.<br>

The method returns a new string that is padded with the specified character to the specified width.<br>

Using :+ to show the sign of a number<br>
Explaination: The :+ format specifier is used to show the sign of a number.<br>
It adds a plus sign (+) before positive numbers and a minus sign (-) before negative numbers.<br>
The method returns a new string representing the number with the appropriate sign.<br>
This format specifier is useful for displaying numbers in a way that clearly indicates their sign, especially in financial or mathematical contexts.<br>
For example, you might use it to format a number as a signed value in a report or a table.<br>
